 # **Tensorflow** 
- A tensor is a generalization of vectors and matrices and is easily understood as a multidimensional array
- A tensor with rank 1 is a one-dimensional array
- a rank-2 tensor can be expressed as a square matrix
- all operations performed are element-wise 

```python
%tensorflow_version 2.x
import tensorflow as tf
print(tf.version)
```

# **Creating Constants**
- once assigned they cannot be changed 

```python
a = tf.constant(2) 
b = tf.constant(3)
```

```python
#in tf2 we do not have to create a session 
tf.print(a) 
```

```python
#sess.run(a) no session.run in tf2
```

```python
c = a+b
tf.print(c)
```

```python
ar1 = tf.constant([[3,3]])
ar2 = tf.constant([[3],[3]])
tf.print(tf.matmul(ar1,ar2)) #Matrix Multiplication 
```

```python
x = tf.constant([1,2,3])
y = tf.constant([4,5,6])
tf.add(x,y)
tf.multiply(x,y)
tf.divide(x,y)
```

```python
#slicing and indexing
x = tf.constant([[1,2,3], [4,5,6]])
tf.print(x[1][2])
#converting to numpy
x = x.numpy()
print(type(x))

#convetring to tensor
x = tf.convert_to_tensor(x)
print(type(x))

#string tensor
string =  tf.constant("AD")
print(string)
```
