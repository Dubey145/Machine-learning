a part of pre processing data 
- it says that "lets get all our data in a uniform range"


what is the impace of having different ranges for different features?
- in most of the algos the feature with the larger range will completely overpower the feature having small ranges

one method to do feature scaling is min max scaling
- if we want our elements to be between 0-1 just divide by max element
- Correct formula for Min Max Scaling is – (Xi – Xmin) / (Xmax – Xmin).

```python
from sklearn import preprocessing
import numpy as np
```

```python
x = np.array([[1.,-1.,2.],
              [2.,0.,0. ],
              [0.,1.,-1.]])
```

```python
scaled_x=preprocessing.scale(x)
print(scaled_x)
print(scaled_x.mean(axis=0))
print(scaled_x.std(axis=0))


#centers data around mean so mean = 0 and variance = 1. 
```

the only problem with scale function is that
- as we will have 2 sep data viz train and test, we will have to combine both and scale it 
- otherwise they will be scaled differently

what we will actually do is that 
- given any training data, we will scale this training data and store the parameters of scaling
- so that when we scale out TESTING data we use the same scaling parameters
- this cannot be done with the regular .scale() function

```python
#we use preprocessing.StandardScaler() we can pass customised param for mean,variance etc
new_x = preprocessing.StandardScaler()
new_x.fit(x) #.fit() will find scaling parameters
new_x.transform(x)
```

```python
x_test = [[1,1,0]]
new_x.transform(x_test)
```
