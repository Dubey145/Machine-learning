how do we find how good or bad our classification model?

1. **Accuracy**
    - but acuraccy isnt always perfect because what if the data was skewed ie 95% data was in class 0 and 5% in class 1 and the model always predicted class 0 for any input given, we will get 95% accuracy but the classifier doesn't actually work
2. **Using confusion matrix**
    - confusion matrix tells us what are the two classes between which confusion is taking place 


```python
from sklearn import datasets
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
iris = datasets.load_iris()
iris.DESCR
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(iris.data,iris.target)
```

```python
algo = LogisticRegression()
algo.fit(xtrain,ytrain)
```

```python
ytrain_pred = algo.predict(xtrain)
```

```python
ytest_pred = algo.predict(xtest)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(ytrain,ytrain_pred)
```

```python
confusion_matrix(ytest,ytest_pred)
```

<h2>Precision and Recall</h2>


precision - what % of predicted is correct <br>
recall - values that were recalled 

```python
from sklearn.metrics import classification_report
print(classification_report(ytest,ytest_pred))
```
