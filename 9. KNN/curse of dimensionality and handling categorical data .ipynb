# Curse Of Dimentionality

- we cannot use all the features in out data set thinking that it will lead to highdr accuracy
- it can also lead to more computation

to address this issue we can
1. assign weights to our features 


2. use gradient descent if we can come up with a cost function
3. Using Feature Selection : selection some and dropping some features 
- this is done through backward elimination ie for each feature find the accuracy by keeping this feature and find the accuracy by not keeping this feature, if the accuracy improves keep it otherwise drop it 

- feature selection is very  useful and score can be imporved easily 

# Handling Categorical Data

categorical data is the labelled data ie yes no, high low etc etc

it becomes difficult to find the distance between 2 datapoints <br>
handling binary data ie yes no is easy as we can put 0 and 1 


if we give data with labels say red green blue some number say 0 1 2 respectively<br>
then what we are effectively saying is that blue is far from red and green is closer to red 

if labels are low mid high then we can assign 0 1 2 as it makes sense <br>
as this is a label with some natural order
