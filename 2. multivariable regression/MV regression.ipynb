```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
import pandas as pd 
```

```python
boston = datasets.load_boston()
```

```python
x = boston.data
y = boston.target
```

```python
data = pd.DataFrame(x)
data
```

```python
data.columns = boston.feature_names
data['age^2'] = data.AGE * data.AGE
data['crim^2'] = data.CRIM **2
data['zn^2'] = data.ZN **2
data['indus^2'] = data.INDUS **2
data.head()
```

```python
x2 = data.values
```

```python
from sklearn import model_selection 
xtrain,xtest,ytrain,ytest = train_test_split(x,y,random_state = 0)
x2train,x2test,y2train,y2test = train_test_split(x2,y,random_state = 0)
#making sure both algos get the same training and test split 
```

```python
from sklearn.linear_model import LinearRegression
algo1 = LinearRegression()
algo2 = LinearRegression()
```

```python
algo1.fit(xtrain,ytrain)
algo2.fit(x2train,y2train)
```

```python
ypred = algo1.predict(xtest)
y2pred = algo2.predict(x2test)
train1_score = algo1.score(xtrain,ytrain)
test1_score = algo1.score(xtest,ytest)
print(train1_score, test1_score)


train2_score = algo2.score(x2train,y2train)
test2_score = algo2.score(x2test,y2test)
print(train2_score, test2_score)
#slightly improved 
```
