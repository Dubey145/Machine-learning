What is Gradient Descent and how can we use it to find optimal coefficients?

to make sure our line fits the data is the most optimal manner, the cost function has to be optimised
- so gradient descent is actualy an optimization algorithm

GD says lets start with a point, calculate the slope and then decide how to minimise it  

what is the coefficient alpha?

the coefficient alpha controls the rate at which we are moving to the minima 
it is called the "learning rate"

if alpha is too high we might overshoot, if its too low we will move slowly

we usually use adaptive alpha which starts with a high value of alpha and then reduces it as we move towards the minima
