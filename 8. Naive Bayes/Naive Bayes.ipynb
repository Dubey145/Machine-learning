# Bayes Theorem


1. in classification given an input x, we have to classify if it is in class a b or c 
2. we calculate the probability p(y=a/x=x), p(y=b/x=x) and p(y=c/x=x)
3. which are conditional probabilities i.e what is the probability that y = a if x is the input (one event of giving input has happened)
4. the class with the max probability is selected as the class of x
