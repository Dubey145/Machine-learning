```python
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.tree import export_graphviz
```

```python
iris = datasets.load_iris()
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(iris.data, iris.target,random_state = 1)
#random state = 1 means that ill get the same split everytime i run this cell 
```

```python
clf = DecisionTreeClassifier()
clf.fit(xtrain,ytrain) #this is where the decision tree is built 
```

```python
import pydotplus
#exporting the tree as pdf
dot_data = export_graphviz(clf,out_file = None,feature_names = iris.feature_names) # outfile = None means that we will not be exporting data instrad the data will be returned here  
#converting into pdf
graph = pydotplus.graph_from_dot_data(dot_data)
graph.write_pdf("iris.pdf")
#pydotplus installation pending 
```

```python
ytrain_pred = clf.predict(xtrain)
ytest_pred = clf.predict(xtest)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(ytrain,ytrain_pred)
#if you have a really good training dataset accuracy the model might be overfitting 
#check with the test data
```

```python
confusion_matrix(ytest,ytest_pred)
```
