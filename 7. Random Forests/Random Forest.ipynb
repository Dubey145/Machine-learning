# Random Forest
- in random forest there are multiple trees 
- the input is fed to multiple trees and the output is noted for each tree
- in the end we go with the majority 

- we use random forests because in sk learn pruning is not implemented
- we can give the depth of the tree in sklearn but we can't prun

- in Random forests the trees are built differently 
- there is randomness in building the trees 



<pre>
Here something is either a or b 
the input is given to 3 trees and 2 predict it to be b and the other thinks it's a
we go with the majority and classify it as b
</pre>

# Regression using DT and Random Forests
- earlier in DT and RF things were either "a" or "b" here in regression the values are continuous so we have to split on a metric
- Usually that metric would be mean squared error (MSE) 
- select the split that has the least mse 
- we find the feature splitting on which mse is minimized 
- Predicted value is the Mean of all samples that belong to that node.




Formula for Mean Squared error should be divided by N (total number of dataset points).
